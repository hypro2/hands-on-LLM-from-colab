```python
from liteparse import LiteParse

parser = LiteParse()

result = parser.parse("/content/sample_data.pdf",
                      ocr_enabled=True,
                      ocr_server_url="http://localhost:8828/ocr",
                      ocr_language="ko",
                      dpi=300)

# Full document text
print(result.text)

# Per-page data
for page in result.pages:
    print(f"Page {page.pageNum}: {len(page.textItems)} text items")
```
